# Compound Interest Flowchart
![compoundinterest](compoundinterest.png)

```python
p= float(input("enter principle"))
r= int(input("enter rate"))
n= int(input("enter number"))
t= int(input("enter time"))
x = n * t
a= p * (int(1+r/n) ^ x)
print ("ci is: ", a)
```

# Annual Plan Flowchart
![annualplan](annualplan.png)

```python
print("calculate annuity plan")
p= float(input("enter principle"))
r= int(input("enter rate"))
n= int(input("enter num"))
t= int(input("enter num"))
m= int(input("enter num"))
a= p*m*t*((int(1+r/n)*n^t-1)/(r/n))
print ("annuity plan is: ", a)
```
